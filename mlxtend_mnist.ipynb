# Data Load

## image data as df

```python
mnist_array = np.loadtxt("mnist_8x8_image.csv", delimiter=',')
mnist_array = mnist_array.astype(np.int64)
col_names = [str(i) for i in (range(mnist_array.shape[1]))]
mnist_df = pd.DataFrame(mnist_array, columns=col_names)
mnist_df.head()
```

## label data as df

```python
mnist_label = np.loadtxt("mnist_8x8_label_onehot.csv", delimiter=",")
col_names = ['label_'+str(i) for i in range(mnist_label.shape[1])]
mnist_label = pd.DataFrame(mnist_label, columns=col_names)

mnist_label.head()
```

# Generating frequent itemsets

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, fpcommon, fpmax
```

```python
%time frequent_itemsets = apriori(mnist_df, min_support=0.14, use_colnames=True)
print(frequent_itemsets.shape)
frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)
frequent_itemsets.head(10)
```

```python
%time frequent_itemsets = fpgrowth(mnist_df, min_support=0.14, use_colnames=True)
print(frequent_itemsets.shape)

frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)
frequent_itemsets.head(10)
#frequent_itemsets.tail(10)
```

```python
%time frequent_itemsets = fpmax(mnist_df, min_support=0.14, use_colnames=True)
print(frequent_itemsets.shape)

frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)
frequent_itemsets.head(10)
frequent_itemsets.tail(10)
```
